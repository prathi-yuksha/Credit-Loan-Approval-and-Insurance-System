```python
encoder = LabelEncoder()
data['Gender']=encoder.fit_transform(data['Gender'])
data['Occupation']=encoder.fit_transform(data['Occupation'])
data['InsuranceType']=encoder.fit_transform(data['Insurance_Type'])
encoded_columns = ['Gender','Occupation','InsuranceType']
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```

```python
def recommend_insurance(customer_data, model, encoder, encoded_columns):
    customer_data_encoded = pd.DataFrame()
    for col in customer_data.columns:
        if col in encoded_columns:
            customer_data_encoded[col] = encoder.fit_transform(customer_data[col])
        else:
            customer_data_encoded[col] = customer_data[col]
    
    probabilities = model.predict_proba(customer_data_encoded)
    top_3_indices = (-probabilities).argsort()[:, :3]
   
    
    return top_3_indices
```

```python
new_customer_data = pd.DataFrame({
    'Age': [35],
    'Gender': ['Female'],
    'Occupation': ['Teacher'],
    'Income': [54000],
    'Premium_Amount': [490]
})
```

```python
recommendations = recommend_insurance(new_customer_data,model, encoder,encoded_columns)
print("Recommended Insurance Types:", recommendations)
```
