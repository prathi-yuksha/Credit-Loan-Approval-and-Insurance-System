```python
testd=test_data.iloc[:,1:11]
testd
```

```python
encoder=LabelEncoder()
var=['Gender', 'Married','Education','Self_Employed']
traind[var]=traind[var].apply(encoder.fit_transform)
traind
```

```python
traind.isna().sum()
```

```python
traind['Credit_History']=traind['Credit_History'].fillna(traind['Credit_History'].mean())
```

```python
traind['Credit_History']=traind['Credit_History'].astype(int)
```

```python
traind['LoanAmount']=traind['LoanAmount'].fillna(traind['LoanAmount'].mean())
```

```python
testd['Credit_History']=testd['Credit_History'].fillna(testd['Credit_History'].mean())
testd['Credit_History']=testd['Credit_History'].astype(int)
```

```python
x=traind[['Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount']]
y=traind['Credit_History']
```
